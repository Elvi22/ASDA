```python
import pandas as pd

# Base URL format for CSV export from Google Sheets
base_url = 'https://docs.google.com/spreadsheets/d/{file_id}/export?format=csv&gid='

# Replace with the actual file ID of the Google Sheets document
file_id = '1ecopK6oyyb4d_7-QLrCr8YlgFrCetHU7-VQfnYej7JY'

# Dictionary to map sheet names to their respective city, country, day_type, and gid
sheet_info = {
    'amsterdam_weekdays': ('Amsterdam', 'Netherlands', 'weekday', '1813557111'),
    'amsterdam_weekends': ('Amsterdam', 'Netherlands', 'weekend', '1671143450'),
    'athens_weekdays': ('Athens', 'Greece', 'weekday', '1474041232'),
    'athens_weekends': ('Athens', 'Greece', 'weekend', '993539767'),
    'berlin_weekdays': ('Berlin', 'Germany', 'weekday', '305266500'),
    'berlin_weekends': ('Berlin', 'Germany', 'weekend', '295085637'),
    'barcelona_weekdays': ('Barcelona', 'Spain', 'weekday', '1009970054'),
    'barcelona_weekends': ('Barcelona', 'Spain', 'weekend', '1739828988'),
    'budapest_weekdays': ('Budapest', 'Hungary', 'weekday', '1438277999'),
    'budapest_weekends': ('Budapest', 'Hungary', 'weekend', '1326018133'),
    'lisbon_weekdays': ('Lisbon', 'Portugal', 'weekday', '82202766'),
    'lisbon_weekends': ('Lisbon', 'Portugal', 'weekend', '1355881051'),
    'london_weekdays': ('London', 'England', 'weekday', '9179440'),
    'london_weekends': ('London', 'England', 'weekend', '58206735'),
    'paris_weekdays': ('Paris', 'France', 'weekday', '1452107999'),
    'paris_weekends': ('Paris', 'France', 'weekend', '1803477090'),
    'rome_weekdays': ('Rome', 'Italy', 'weekday', '1127044053'),
    'rome_weekends': ('Rome', 'Italy', 'weekend', '1238155247'),
    'vienna_weekdays': ('Vienna', 'Austria', 'weekday', '671930913'),
    'vienna_weekends': ('Vienna', 'Austria', 'weekend', '1950822222')
}

# List to store individual dataframes
dfs = []

# Loop through each sheet in the sheet_info dictionary
for sheet_name, (city, country, day_type, gid) in sheet_info.items():
    # Construct the full URL for each sheet
    sheet_url = f'https://docs.google.com/spreadsheets/d/{file_id}/export?format=csv&gid={gid}'
    try:
        # Load the data into a DataFrame
        df = pd.read_csv(sheet_url)
        
        # Add city, country, and day_type columns
        df['city'] = city
        df['country'] = country
        df['day_type'] = day_type
        
        # Append the DataFrame to the list
        dfs.append(df)
        print(f"Sheet '{sheet_name}' loaded successfully.")
        
    except Exception as e:
        print(f"Failed to load sheet '{sheet_name}': {e}")

# Merge all DataFrames into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# Rename 'realSum' column to 'Price' if it exists
if 'realSum' in merged_df.columns:
    merged_df = merged_df.rename(columns={'realSum': 'Price'})

# Display the merged DataFrame
merged_df
```

Is there a difference in price between two cities? Choose at least three pairs of cities during weekdays.

```python
from scipy.stats import ttest_ind
# Filter data for weekday
weekday_data = merged_df[merged_df['day_type'] == 'weekday']


# List of pairs of cities to compare
city_pairs = [('Amsterdam', 'Berlin'),('Rome', 'Barcelona'),('Lisbon', 'Budapest')]

# Conduct t test
for city1,city2 in city_pairs:
    price_city1 = weekday_data[weekday_data['city'] == city1]['Price']
    price_city2 = weekday_data[weekday_data['city'] == city2]['Price']
    # perform the test
    t_stat, p_value = ttest_ind(price_city1, price_city2, equal_var=False)
    
    print(f"T-test between {city1} and {city2}:")
    print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Result: Significant difference in prices (p < 0.05)")
    else:
        print("Result: No significant difference in prices (p >= 0.05)")
    print("-" * 40)
```

Does the price tend to be higher on weekends?

```python
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for weekday and weekend
weekday_prices = merged_df[merged_df['day_type'] == 'weekday']
weekend_prices = merged_df[merged_df['day_type'] == 'weekend']
mean_weekday_price = weekday_prices['Price'].mean()
mean_weekend_price = weekend_prices['Price'].mean()

print(f"Mean price for Weekdays: {mean_weekday_price:.2f}")
print(f"Mean price for Weekends: {mean_weekend_price:.2f}")

# perform t test
t_stat, p_value = ttest_ind(weekday_prices['Price'], weekend_prices['Price'], equal_var=False)
print("\nT-test Results")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Result: Significant difference in prices (p < 0.05)")
else:
    print("Result: No significant difference in prices (p >= 0.05)")

                           
# Plot the data for better visualization
plt.figure(figsize=(12, 6))
sns.boxplot(x='day_type', y='Price', data=merged_df, palette="pastel")
plt.title('Comparison of Prices on Weekdays vs Weekends')
plt.xlabel('Day Type')
plt.ylabel('Price')
plt.show()
```

Are listings of superhosts more expensive than those of normal hosts?

```python
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for superhosts and normal hosts
superhost_prices = merged_df[merged_df['host_is_superhost'] == True]['Price']
normal_host_prices = merged_df[merged_df['host_is_superhost'] == False]['Price']
mean_superhost_price = superhost_prices.mean()
mean_normal_host_price = normal_host_prices.mean()
print(f"Mean price for Superhosts: {mean_superhost_price:.2f}")
print(f"Mean price for Normal Hosts: {mean_normal_host_price:.2f}")

# Perform t test 
t_stat, p_value = ttest_ind(superhost_prices, normal_host_prices, equal_var=False)
print("\nT-test Results")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Result: Superhost listings are significantly more expensive (p < 0.05)")
else:
    print("Result: No significant difference in price between superhost and normal host listings (p >= 0.05)")

# Plot the data
plt.figure(figsize=(12, 6))
sns.boxplot(x='host_is_superhost', y='Price', data=merged_df, palette="pastel")
plt.title('Comparison of Prices between Superhosts and Normal Hosts')
plt.xlabel('Host Type')
plt.ylabel('Price')
plt.show()
```

Are superhosts closer to the city center and metro station than normal hosts

```python
from scipy.stats import ttest_ind

# Filter data into superhost and normal host groups
superhost_data = merged_df[merged_df['host_is_superhost'] == True]
normal_host_data = merged_df[merged_df['host_is_superhost'] == False]
# Calculate average distances for both groups
mean_dist_center_superhost = superhost_data['dist'].mean()
mean_dist_center_normal = normal_host_data['dist'].mean()
mean_dist_metro_superhost = superhost_data['metro_dist'].mean()
mean_dist_metro_normal = normal_host_data['metro_dist'].mean()

print(f"Average Distance to City Center for Superhosts: {mean_dist_center_superhost:.2f}")
print(f"Average Distance to City Center for Normal Hosts: {mean_dist_center_normal:.2f}")
print(f"Average Distance to Metro Station for Superhosts: {mean_dist_metro_superhost:.2f}")
print(f"Average Distance to Metro Station for Normal Hosts: {mean_dist_metro_normal:.2f}")

# Perform t-tests for distances to the city center and metro station
t_stat_center, p_value_center = ttest_ind(superhost_data['dist'], normal_host_data['dist'], equal_var=False)
t_stat_metro, p_value_metro = ttest_ind(superhost_data['metro_dist'], normal_host_data['metro_dist'], equal_var=False)

# Display t-test results
print("\nT-test Results for Distance to City Center")
print(f"T-statistic: {t_stat_center:.2f}, P-value: {p_value_center:.4f}")
if p_value_center < 0.05:
    print("Result: Superhosts are significantly closer to the city center (p < 0.05)")
else:
    print("Result: No significant difference in distance to the city center (p >= 0.05)")

print("\nT-test Results for Distance to Metro Station")
print(f"T-statistic: {t_stat_metro:.2f}, P-value: {p_value_metro:.4f}")
if p_value_metro < 0.05:
    print("Result: Superhosts are significantly closer to metro stations (p < 0.05)")
else:
    print("Result: No significant difference in distance to metro stations (p >= 0.05)")
```

Is there a pattern between room types and superhost status?

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a cross-tabulation of room type and superhost status
room_superhost_ct = pd.crosstab(merged_df['room_type'], merged_df['host_is_superhost'])

# Print the cross-tabulation for reference
print("Cross-tabulation of Room Type and Superhost Status:")
print(room_superhost_ct)
# Perform a Chi-squared test
chi2, p, dof, expected = chi2_contingency(room_superhost_ct)

print("\nChi-squared Test Results")
print(f"Chi-squared Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
if p < 0.05:
    print("Result: There is a statistically significant association between room type and superhost status (p < 0.05).")
else:
    print("Result: No significant association between room type and superhost status (p >= 0.05).")
```
